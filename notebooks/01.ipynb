```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def squares():
  A = np.zeros([50, 50])
  A[1:10,1:20] = 1
  A[29:49,1:20] = 1
  A[1:20,29:49] = 1
  A[29:49, 26:49] = 1
  return A

A = squares()
plt.imshow(A)
```

```python
def corners():
  B = np.zeros([50,50])
  for i in range(50):
    B[i,i:] = 1
  A = np.flip(B,axis=1)
  AB = np.hstack((A,B))
  CD = np.flip(AB,axis=0)
  return np.vstack((AB,CD))

A = corners()
plt.imshow(A)
```

```python
def scotland():
  A = np.eye(50)
  for i in range(50):
    A[-i-1, -i] = 1
    A[-i-1, i] = 1
    A[-i, i] = 1
    A[i, -i] = 1
    A[i, -i-1] = 1
    A[-i, -i-1] = 1
  return A

A = scotland()
plt.imshow(A)
```
